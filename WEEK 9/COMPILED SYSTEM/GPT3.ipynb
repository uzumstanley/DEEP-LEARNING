```python
from openai import OpenAI
client = OpenAI(api_key = "")


completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
    {"role": "user", "content": "Compose a poem that explains the concept of recursion in programming."}
  ]
)

print(completion.choices[0].message.content)
```

```python

def summarize_text(input_text):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
         {"role": "user", "content": f"Summarize the following text:\n{input_text}"} ])
    return response.choices[0].message.content

def generate_questions_and_answers(input_text, num_questions=5):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
         {"role": "user", "content": f"Generate {num_questions} questions and answers based on the following text:\n{input_text}"} ])
    return response.choices[0].message.content
    #return [qa['text'] for qa in response.choices[0]]

if __name__ == "__main__":
    # Example text to be summarized and used for question generation
    example_text = """
    The internet has become an essential part of our daily lives. It has revolutionized communication, commerce, and education.
    With the rise of social media, people can connect with others around the world. E-commerce has allowed businesses to reach
    a global market. Online education has made learning accessible to people everywhere. However, the internet also poses
    challenges such as privacy concerns and the spread of misinformation.
    """

    # Summarize the text
    summarized_text = summarize_text(example_text)
    print("Summarized Text:")
    print(summarized_text)
    print("\n")

    # Generate questions and answers
    generated_qas = generate_questions_and_answers(summarized_text)
    print("Generated Questions and Answers:")
    #for i, qa in enumerate(generated_qas, start=1):
    #    print(f"Q{i}: {qa['question']}")
    #    print(f"A{i}: {qa['answer']}")
    #    print("\n")
    print(generated_qas)
```

This code will use the OpenAI API to access a GPT-3 model and generate text based on the given prompt. The generated text will be a test to evaluate the performance of a machine learning model that has been trained to classify images of animals.



```python
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
    {"role": "user", "content": """Write a test to evaluate the performance of a machine learning model.that has been trained to classify images of animals.
    The test should include at least 10 images, each of which is labeled with the correct animal species.
     The test should also include a set of images that are not labeled, and the model should be required to classify these as well.
     Finally, the test should calculate the model's accuracy on the test set and print out the results."""} ])



# Print the generated text

print(completion.choices[0].message.content)
```

### **`Grammar correction`**
https://beta.openai.com/examples/default-grammar

```python
response = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "system",
      "content": "You will be provided with statements, and your task is to convert them to standard English."
    },
    {
      "role": "user",
      "content": "She no went to the market."
    }
  ],
  temperature=0.7,
  max_tokens=64,
  top_p=1
)
print(response.choices[0].message.content)
```
